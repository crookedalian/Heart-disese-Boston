#. At the end section some acuracy and confusion matrics and the pridiction results


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
test_missing = pd.read_csv('C:/Users/Ali/miniconda3/numpy/heart-disease-artifitially-missing-data.csv')
test_missing.head()
```

```python
test_missing.dtypes
```

```python
test_missing["thalach"]=pd.to_numeric(test_missing["thalach"], errors='coerce')
```

```python
test_missing.dtypes
```

```python
print(test_missing.isna())
```

```python
print(test_missing.isnull().sum())
```

```python
test_missing_imputed = test_missing.fillna(test_missing.mean())
```

```python
X=test_missing.drop("target",axis=1)
y=test_missing["target"]
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
y_preds=clf.predict(X_test)
y_preds.shape
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
#Try different amounts of n_estimaters
np.random.seed(42)
for i in range(10,100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model accuracy on test set: {clf.score(X_test,y_test)* 100:0.2f}")
    print("")
```

```python
#save a model and load it
import pickle
pickle.dump(clf,open("random_forest_model_1.pk1","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1","rb"))
loaded_model
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_preds)
cm = confusion_matrix(y_test, y_preds)
cr = classification_report(y_test, y_preds)

# Print evaluation results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
print("Classification Report:\n", cr)
```
